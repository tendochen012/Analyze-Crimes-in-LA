```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
#Peak Crime Hour
crimes["HOUR"] = crimes["TIME OCC"].apply(lambda x: x[:2])
peak_crime_hour = int(crimes["HOUR"].value_counts()[crimes["HOUR"].value_counts() == crimes["HOUR"].value_counts().max()].index[0])
display(peak_crime_hour)
```

```python
crimes["HOUR"] = crimes["HOUR"].astype("category")
night_crime = crimes[crimes["HOUR"].isin(["22","23","00","01","02","03"])]
peak_night_crime_location = night_crime["AREA NAME"].value_counts()[night_crime["AREA NAME"].value_counts() ==night_crime["AREA NAME"].value_counts().max()].index[0]
display(peak_night_crime_location)
```

```python
#Identiy the numer of crimes committed against victims by age group


#Conditions for putting the bucket correctly

age_bins = [0,17,25,34,44,54,64,np.inf]
age_labels = ["<18", "18-25", "26-34", "35-44","45-54", "55-64", "65+"]

crimes["Age_Group"] = pd.cut(crimes["Vict Age"],
                               bins=age_bins,
                               labels=age_labels)

#Data Aggregation based on Age_Group and Crime_ID

#Desired Series
victim_ages = crimes["Age_Group"].value_counts()
display(victim_ages)
display(crimes)
```
